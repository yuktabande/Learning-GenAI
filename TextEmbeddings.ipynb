```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import numpy as np

# Make sure text_embedding is a NumPy array
text_embedding = np.array(text_embedding)

# Apply similarity calculation
df['similarity score'] = df['Embeddings'].apply(
    lambda x: cosine_similarity(np.array(x).reshape(1, -1), text_embedding.reshape(1, -1))[0][0]
)
```
